```python
from sklearn import datasets

# Load the iris dataset from sklearn
# https://scikit-learn.org/1.4/modules/generated/sklearn.datasets.load_iris.html#sklearn.datasets.load_iris
iris = datasets.load_iris(as_frame=True)
```

```python
type(iris)  # This is not dataframe
```

```python
type(iris.data)  # This is dataframe
```

```python
# Show the first 5 rows of the DataFrame.
irisdata = iris.data
irisdata.head(5)
```

```python
# Show the column names.
irisdata.columns
```

```python
# Show the data types of each column.
irisdata.dtypes
```

```python
# Show the summary statistics of the DataFrame.
irisdata.describe()
```

```python
# Count the number of rows in the DataFrame.
len(irisdata)
```

```python
# Count the number of unique values in a specific column.
for col in irisdata.columns:
  print(f"The number of unique values in '{col}': {len(irisdata[col].unique())}")
```

```python
# Sort the DataFrame by a specific column.
irisdata.sort_values(by='sepal length (cm)')
```

```python
# Filter the DataFrame based on a condition. - sepal length is 4.4 and petal length is 1.3
irisdata.query('`sepal length (cm)` == 4.4 and `petal length (cm)` == 1.3')
```

```python
# Filter the DataFrame based on a condition. -
irisdata[irisdata['sepal width (cm)'] > 3.0].sort_values(by='sepal width (cm)')
```

```python
# Group the DataFrame by a specific column and calculate the mean of another column.
irisdata.groupby('sepal length (cm)')['sepal width (cm)'].mean()
```

```python
# Add a new column to the DataFrame.
irisdata['sepal ratio'] = irisdata['sepal length (cm)']/irisdata['sepal width (cm)']
irisdata.head()
```

```python
# Write a csv (little clunky in google colab)
from google.colab import files
irisdata.to_csv('mcmurtrie_lab1_iris_data.csv', index=False)
files.download('mcmurtrie_lab1_iris_data.csv')
```
